```python
import pickle
import numpy as np
import pandas as pd
import gzip
from zipfile import ZipFile
import os 
from glob import glob
```

Collect all pkl

```python
synth_df = pd.DataFrame()

with ZipFile('intermediate_files/results_synth.zip', 'r') as zip:
    files = [file.orig_filename for file in zip.filelist]
c=0
for res in files:
    with ZipFile('intermediate_files/results_synth.zip', 'r') as zip:
        with gzip.open(zip.open(res)) as file:
            dc = pickle.load(file)

        N = dc['N']
        M = dc['M']
        k = dc['k']
        a = dc['a']
        idx = res.split('_')[4][2:]

        for config in list(dc.keys())[5:]:
            dict_to_push = dict.copy(dc[config])
            dict_to_push['algo'] = res.split('_')[-1][:-7]
            dict_to_push['N'] = N
            dict_to_push['M'] = M
            dict_to_push['a'] = a
            dict_to_push['TrueK'] = k
            dict_to_push['id'] = idx

            if config[0] == 'elbow':
                name = f'{config[0]}_{config[1][0]}{config[1][1]}'
            else:
                name = config[0]
            dict_to_push['index'] = name

            if config[0] == 'silhouette':
                dist = '-'
            else:
                dist = config[-2]
            dict_to_push['distance'] = dist
            dict_to_push['agg'] = config[-1]

            synth_df = pd.concat([synth_df, pd.DataFrame(dict_to_push, index=[c])], axis = 0, ignore_index = True, join='outer')
            c+=1
            
synth_df.to_csv('intermediate_files/synth_results_v2.csv',  header=True, index=False)
```

```python
uci_df = pd.DataFrame()
c=0

with ZipFile('intermediate_files/results_uci.zip', 'r') as zip:
    files = [file.orig_filename for file in zip.filelist]

for res in files:
    with ZipFile('intermediate_files/results_uci.zip', 'r') as zip:
        with gzip.open(zip.open(res)) as file:
            dc = pickle.load(file)
        
        if len(res.split('_')) == 4:
            dataset = res.split('_')[0]+res.split('_')[1]
            idx = res.split('_')[2]
        else:
            dataset = res.split('_')[0]
            idx = res.split('_')[1]
        N = dc['N']
        M = dc['M']
        k = dc['k']
        algorithm = res.split('_')[-1][:-7]
        
        for config in list(dc.keys())[-37:]:
            dict_to_push = dict.copy(dc[config])
            dict_to_push['algo'] = algorithm
            dict_to_push['dataset'] = dataset
            dict_to_push['N'] = N
            dict_to_push['M'] = M
            dict_to_push['TrueK'] = k
            dict_to_push['id'] = idx

            if config[0] == 'elbow':
                name = f'{config[0]}_{config[1][0]}{config[1][1]}'
            else:
                name = config[0]
            dict_to_push['index'] = name

            if config[0] == 'silhouette':
                dist = '-'
            else:
                dist = config[-2]
            dict_to_push['distance'] = dist
            dict_to_push['agg'] = config[-1]

            uci_df = pd.concat([uci_df, pd.DataFrame(dict_to_push, index=[c])], axis = 0, ignore_index = True, join='outer')
            c+=1

uci_df.to_csv('intermediate_files/uci_results_v2.csv',  header=True, index=False)
```
